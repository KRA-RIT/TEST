```python
import sys
from filecmp import dircmp

#path1 = r'C:\ISI_DATA\MDVL_AMA-06778'
#path2 = r'C:\ISI_DATA\MDVL_AMA-06778 - Copy'

path1 = r'M:\ISI result\DRIVE\MDVL_RO-01617_PRE'
path2 = r'I:\ISI\MDVL_RO-01617_PRE'
```

```python
def print_diff_files(dcmp):
    for name in dcmp.diff_files:
        print("diff_file %s found in %s and %s" % (name, dcmp.left, dcmp.right))
    if dcmp.left_only:
        print('%s: %s ' % (dcmp.left, dcmp.left_only))
    if dcmp.right_only:
        print('%s: %s ' % (dcmp.right, dcmp.right_only))
    for sub_dcmp in dcmp.subdirs.values():
        print_diff_files(sub_dcmp)
```

```python
dcmp = dircmp(path1, path2)
print_diff_files(dcmp)
```
